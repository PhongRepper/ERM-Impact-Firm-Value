```python
#y = df1['TobinQ']
#X = df1[['ERMI Index', 'SIZE', 'ROA', 'Gearing', 'Growth Rate', 'Dividend or not', 'International Business', 'Beta 2022']]

# Concatenate X and y to ensure that they have matching indices and drop rows with missing values
df_cleaned = pd.concat([y, X], axis=1).dropna()

# Reassign cleaned X and y after removing missing values
y_cleaned = df_cleaned['TobinQ']
X_cleaned = df_cleaned[['ERMI Index', 'ROA', 'SIZE', 'Growth Rate', 'Gearing', 'Beta 2022', 'Dividend or not', 'International Business']]
#X_cleaned = df_cleaned[['ERMI Index', 'ROA', 'SIZE', 'Growth Rate', 'Gearing','Dividend or not']]

# Add constant (intercept) to the independent variables
X_cleaned_with_const = sm.add_constant(X_cleaned)

# Perform the OLS regression
model = sm.OLS(y_cleaned, X_cleaned_with_const).fit()

# Print the summary of the regression
print(model.summary())
```

```python

# Extract coefficients and p-values
coefficients = model.params
p_values = model.pvalues

# Define significance stars
def significance_stars(p):
    if p < 0.01:
        return '***'
    elif p < 0.05:
        return '**'
    elif p < 0.1:
        return '*'
    else:
        return ''

# Create a DataFrame for the results
results_df = pd.DataFrame({
    'Coefficient': coefficients,
    'P-value': p_values
})

# Apply significance stars
results_df['Significance'] = results_df['P-value'].apply(significance_stars)

# Format the coefficients with significance stars
results_df['Coefficient'] = results_df.apply(lambda row: f"{row['Coefficient']:.4f}{row['Significance']}", axis=1)


# Print the formatted results
print(results_df)
```

```python
# Perform the Breusch-Pagan test
bp_test = het_breuschpagan(model.resid, model.model.exog)
```

```python
# bp_test returns a tuple: (Lagrange multiplier statistic, p-value, f-value, f p-value)
lm_statistic, lm_pvalue, f_statistic, f_pvalue = bp_test

print("Breusch-Pagan test statistic:", lm_statistic)
print("p-value:", lm_pvalue)
print("F-statistic:", f_statistic)
print("F-test p-value:", f_pvalue)
```

--> No heteroskedasticity

```python
# Extract residuals from the model
residuals = model.resid

# Perform the Shapiro-Wilk test on the residuals
statistic, p_value = stats.shapiro(residuals)

print("Shapiro-Wilk test statistic:", statistic)
print("p-value:", p_value)
```
